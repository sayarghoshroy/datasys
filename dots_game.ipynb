```python
# Dots Game
N = 4
M = 5

score_A = 0
score_B = 0

# N, M: Number of dots in each Row and each Column Respectively
```

```python
board_filled_index = {}
# if stores drawn lines
# line: Top Left: (x, y), Type: 1 (Horizontal), 2 (Vertical)
# maps the marked line to the move number
# i.e the actual serial number of the move

# indexing the board using the least amount of space

one_fills = {}
# stores a line structure which when filled shall complete a block

def check_if_filled(x, y, ltype):
    # check if a move is valid
    # checks whether a line is actually drawn or not
    if x not in board_filled_index:
        board_filled_index[x] = {}
        board_filled_index[x][y] = {}
        return True
    if y not in board_filled_index[x]:
        board_filled_index[x][y] = {}
        return True
    if ltype not in board_filled_index[x][y]:
        return True
    return False

def check_valid_line(x, y, ltype):
    # check if the move is actually legal
    try:
        x = int(x)
        y = int(y)
        ltype = int(ltype)
    except:
        return False
    if x < 1 or x > M or y < 1 or y > M:
        return False
    
    return True
```

```python
# to fill in the index
# need to give in the move
# and the move serial number in the running game

def fill(x, y, ltype, serial):
    if check_valid_line(x, y, ltype) == True and check_if_filled(x, y, ltype) == True:
        board_filled_index[x][y][ltype] = serial
        print("Move Filled in Successfully")
        return 1
        
    else:
        print("Invalid Move")
        return 0
    
def done(x, y, ltype):
    above = 0
    below = 0
    left = 0
    right = 0
    
    if ltype == 1:
        # horizontal
        # checking box above line
        if x in board_filled_index and (y - 1) in board_filled_index[x] and 1 in board_filled_index[x][y - 1]:
            if x in board_filled_index and (y - 1) in board_filled_index[x] and 2 in board_filled_index[x][y - 1]:
                if (x + 1) in board_filled_index and (y - 1) in board_filled_index[x + 1] and 2 in board_filled_index[x + 1][y - 1]:
                    above = 1
        if x in board_filled_index and (y + 1) in board_filled_index[x] and 1 in board_filled_index[x][y + 1]:
            if x in board_filled_index and y in board_filled_index[x] and 2 in board_filled_index[x][y]:
                if (x + 1) in board_filled_index and y in board_filled_index[x + 1] and 2 in board_filled_index[x + 1][y]:
                    below = 1         
    if ltype == 2:
        # vertical
        # checking on two sides of line
        if (x - 1) in board_filled_index and y in board_filled_index[x - 1] and 1 in board_filled_index[x - 1][y]:
            if (x - 1) in board_filled_index and y in board_filled_index[x - 1] and 2 in board_filled_index[x - 1][y]:
                if (x - 1) in board_filled_index and (y + 1) in board_filled_index[x - 1] and 1 in board_filled_index[x - 1][y + 1]:
                    left = 1
        if x in board_filled_index and y in board_filled_index[x] and 1 in board_filled_index[x][y]:
            if x in board_filled_index and (y + 1) in board_filled_index[x] and 1 in board_filled_index[x][y + 1]:
                if (x + 1) in board_filled_index and y in board_filled_index[x + 1] and 2 in board_filled_index[x + 1][y]:
                    right = 1
                    
    return above + below + left + right
        
def new_game():
    # clears the board
    # initiates a new game
    board_filled_index.clear()
    print("New Game:")
```

```python
moves_made = 0
turn = 0
new_game()

def print_moves():
    print("Moves:")
    for i in board_filled_index.keys():
        for j in board_filled_index[i].keys():
            for k in board_filled_index[i][j].keys():
                print(i, j, k, board_filled_index[i][j][k])

while(moves_made < (N * M * 2)):
    print("Player 1 Score = " + str(score_A))
    print("Player 2 Score = " + str(score_B))
    print("Player " + str(turn + 1) + "'s Move:")
    move_x, move_y, move_type = [int(i) for i in (input().split(" "))]
    
    if fill(move_x, move_y, move_type, moves_made + 1) == 1:
        moves_made += 1
        points = done(move_x, move_y, move_type)
        
        if points > 0:
            # same player can go again
            if turn == 0:
                score_A += points
            else:
                score_B += points
        else:
            turn = (turn + 1) % 2
```
